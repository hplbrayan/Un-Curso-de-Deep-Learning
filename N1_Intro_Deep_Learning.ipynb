<h1><center>Deep Learning</center></h1>

El Deep Learning (DL), o aprendizaje profundo, es un conjunto de algoritmos de Machine Learning (ML) que buscan, a traves del procesamiento por múliples capas, aprender las principales características de un conjunto de datos y a partir de estas realizar predicciones. El número de capas a partir del cual podríamos decir que un algoritmo hace parte de la familia del DL puede variar de autor a autor, pero la gran mayoría pondría el umbral en dos capas ubicadas entre la entrada y la salida de toda la estructura. 

Hablar a profundidad de la historia del DL y, a modo general, de la Inteligencia Artificial, no es el objetivo de este curso, razón por la cual no se mencionará en detalle estos aspectos, solo se hará énfasis en aquellos que se consideren necesarios. Para comenzar, dirémos que el DL y a modo general la AI, están fuertemente inspirados en el funcionamiento del cerebro humano y en especial de su principal estructura conocida como **neurona**.


Fue en el año de $1906$ cuando Santiago Ramón y Cajal ganó el premio nobel en fisiología o medicina por sus aportes al entendimiento de la estructura del cerebro, desde entonces fue considerado como el fundador de *“La doctrina de la Neurona”* y fueron estos trabajos básicos los que cimentarían las bases para los posteriores estudios de Warren McCulloch y Walter Pitts que dieron origen al primer modelo matemático de una neurona conocido como el *Threshold Logic Unit* (TLU), o *Linear Threshold Unit*.


[Imágen](https://www.elespanol.com/omicrono/tecnologia/20170118/ramon-cajal-arte-dibujar-cerebro/186982241_0.html).


[Imágen](https://es.wikipedia.org/wiki/Neurona#/media/Archivo:Complete_neuron_cell_diagram_es.svg).  

La idea es relativamente sencilla pero muy poderosa. Una neurona recibe estimulos de su entorno a traves de sus *dendrítas*, estos son procesados en su *núcleo* y finalmente, si los impulsos son lo suficientemente fuertes entonces desencadenan una respuesta por parte de la neurona que viaja por su *axon*, estimulo que acabará por activar otras neuronas o algún otro tipo de estructura. Existe un umbral para la magnitud de este estímulo conocido como [potencial de acción](https://es.wikipedia.org/wiki/Potencial_de_acci%C3%B3n), si la combinación de todas las señales que entran por las dendritas no produce un potencial eléctrico superior al potencial de acción entonces la neurona no producirá respuesta alguna. 


[Imagen](https://es.wikipedia.org/wiki/Potencial_de_acci%C3%B3n). 

Esta idea, sumada a la [regla de Hebb](https://es.wikipedia.org/wiki/Teor%C3%ADa_hebbiana):

>*Cuando el axón de una célula A está lo suficientemente cerca de una célula B como para excitarla y participa repetida o persistentemente en su disparo, ocurre algún proceso de crecimiento o cambio metabólico, en una o ambas células, de tal modo que la eficacia de A en disparar a B se ve aumentada*.

proporcionaron la base para instaurar el primer algoritmo de aprendizaje de una neurona artificial (se aclara que la siguiente notación y términos se irán abordando a lo largo de todo el curso).

\begin{equation}
w_{i}^{j+1} = w_{i}^{j} - \alpha x_iy
\end{equation}

donde $w_{i}^{j}$ es el peso de la $i$-ésima entrada a la neurona, $\alpha$ se conoce como la tasa de aprendizaje y, por definición, es positiva, $x_i$ es la $i$-ésima entrada a la neurona, $y$ es la salida de la neurona y $j$ es la $j$-ésima época, o iteración en el proceso de actualización de los pesos. Con esta notación podemos realizar el simil entre una neurona real y una artificial como sigue:

## Arquitectura básica de una neurona artificial. 

Inspirada en la neurona biológica, la neurona artificial está compuesta por entradas (las features) $x_i$, los pesos $w_i$, una suma ponderada de las features por los pesos $z=\sum x_iw_i$, una función de activación $f(z)$ y la salida o respuesta $y$. A continuaión daremos una breve descripción de las principales componentes de la arquitectura de la neurona:

* **Las features $x_i$**: Como se ha visto en el curso de ML, las features son cada una de las variables
    que utiliza el modelo para realizar la predicción.
* **Los pesos $w_i$**: También conocidos como pesos sinápticos, determinan la importancia de cada una de las 
    features del modelo y su determinación es el objetivo de cualquier algoritmo de DL. Cada entrada a la neurona,
    es decir, cada feature, será multiplicada por un peso y será este quien determine el impacto de dicha 
    característica sobre la predicción final (cuando solo tenemos una neurona). A modo general,
    cuando hablamos de algoritmos con muchas capas es la combinación de múltiples pesos lo que determinará
    el impacto de cada feature, no el de uno solo.
* **Suma ponderada $z$**: Lo que determina el estado de activación de la neurona son dos cosas, la suma
    ponderada y la función de activación. $z$ representa la combinación de todas las features pesadas por su 
    importancia (es decir, los pesos) dando como resultado un único valor que será utilizado por la función de 
    activación para determinar la salida de la neurona.
* **Función de activación $f$**: Es la regla que determina el valor que finalmente saldrá de la neurona, utiliza
    como argumento la suma ponderada $z$ para dar el estado final de la misma. Por ejemplo, una de las primeras
    funciones de activación propuestas es la función escalón, definida de manera que $f(z<0.5)=0$ y $f(z\ge 0.5)=1$,
    esta está fuertemente inspirada en la idea del potencial de acción, utilizando como umbral $0.5$. Estos
    dos valores, $0$ y $1$, o estados de la neurona, se pueden utilizar para resolver algunos problemas de 
    clasificación binaria. 
    
Estos son algunos de los términos básicos que se deben entender para poder comprender el mecanismo de funcionamiento
de una simple neurona y, posteriormente, de toda una red neuronal. Como se dijo anteriormente, el objetivo de cualquier algoritmo de DL es el de determinar el valor de los pesos $w_i$ que proporcionan la mejor respuesta posible por parte de la neurona (o red de neuronas). A continuación se estudiará el del que se conoce como la primera neurona artificial, el **perceptrón**.

## El perceptrón

Propuesto por Frank Rosenblatt, el perceptrón es el modelo matemático más sencillo que podamos tener para representar una neurona. 



Este, tal y como fue propuesto inicialmente, solo puede resolver problemas de clasificación binaria que son separables linealmente, es decir, si existe un plano $n$-dimensional que separe totalmente las clases. El algoritmo del perceptrón es:

\begin{equation}
w_{i}^{j+1} = w_{i}^{j} - \alpha x_i(\delta - y)
\end{equation}

donde 
* $w_{i}^{j}$ es el peso de la $i$-ésima entrada a la neurona.
* $\alpha$ es la tasa de aprendizaje positiva.
* $x_i$ es la $i$-ésima entrada a la neurona.
* $y$ es la salida de la neurona.
* $\delta$ es la salida esperada de la neurona, diferente a $y$ que es la que realmente sale de ella.
* $j$ representa la $j$-ésima época o iteración en el algoritmo que calcula los pesos. 

Es importante darse cuenta de la relación existente entre este algoritmo y la regla de Hebb. De hecho, el algoritmo del perceptrón representa una mejora a la regla de Hebb, pues la variación en los pesos sinápticos viene dada de manera proporcional por la diferencia entre lo que la neurona predice y aquello que debería predecir. 
Si $\delta=y$ entonces $\delta-y=0$ lo cual implica que no hay una actualización del peso para esa feature. Si $\delta = 0$ pero $y=1$ entonces $\delta-y=0-1=-1$ de modo que si $x_i>0$ entonces el valor final del peso $w_i$  será disminuido, implicando esto un "castigo" a dicho peso, pero si $x_i<0$ entonces el peso aumentará su valor de manera que se le "premia" por estar en la misma "dirección" que el valor de la feature. Similar argumento se puede realizar en el caso $\delta=1$ y $y=0$.  

Como se dijo, el perceptrón sólo puede resolver problemas que son separables linealmente, es decir, un problema como el de la compuerta XOR, es imposible de resolver con un único perceptrón. Los problemas de las compuertas AND, OR, por otro lado, sí se pueden resolver con el perceptrón.


[Imagen](https://pythonmachinelearning.pro/perceptrons-the-first-neural-networks/).

A parte del problema de la linealidad, otro problema del perceptrón es que, si bien es capas de encontrar la frontera de decisión para clases separables linealmente, la frontera encontrada no es necesariamente la más óptima, de hecho, los pesos finales encontrados por el perceptrón dependen de los valores con los cuales estos se inicializan, llevando esto a soluciones carentes de unicidad. Para ejemplificar el algoritmo del perceptrón a continuación se crea la clase Perceptrón.

## Creando un Perceptrón

```python
class Perceptron:
    
    #====================================================================#
    #                          init the class                            #
    #====================================================================#
    def __init__(self,
                 alpha=0.1,
                 threshold=0.5,
                 max_epochs=100,
                 random_state=False):
        
        # learning ratio alpha
        self.alpha = alpha
        
        # threshold
        self.threshold = threshold
        
        # epochs
        self.max_epochs = max_epochs
        self.epochs = 0
        
        # random state
        if random_state == 0 and (type(random_state)==int or type(random_state)==float):
            random_state = True
            self.random_state = random_state
        else:
            self.random_state = random_state
            
        # convergency state
        self.succes = "--"
        
            
    #====================================================================#
    #                        weigthed sum of features                    #
    #====================================================================#
    def _weighted_sum(self, xs, ws):
        return sum([xs[i]*ws[i] for i in range(self.cols)])
    
    
    #====================================================================#
    #                              Perceptron                            #
    #====================================================================#
    def perceptron(self, xs, ds, ws="auto"):
        
        # weights
        self.cols = len(xs[0])
        self.rows = len(xs)
        self.xs = xs
        self.ds = ds
        
        if ws=="auto":
            self._ws = ws
            self.ws = [0]*self.cols
            
        elif ws=="random":
            import numpy as np
            self._ws = ws
            if self.random_state:
                np.random.seed(self.random_state)
                self.ws = np.ndarray.tolist(
                                            np.random.uniform(low=0,
                                            high=1,
                                            size=self.cols)
                                            )
            else:
                self.ws = np.ndarray.tolist(
                                        np.random.uniform(low=0,
                                        high=1,
                                        size=self.cols)
                                        )
                
            
        elif (type(ws)==list or type(ws)==tuple) and len(ws)==self.cols:
            self._ws = f"lisf of shape {len(ws),}" if type(ws)==list else f"tuple of shape {len(ws),}"
            self.ws = ws
        
        else:
            raise Exception(f"wrong value for ws")
            return
            
            
        # Perceptron algorithm
        WS     = [self.ws]
        errs   = []
        sums   = []
        epochs = 0
        
        while epochs<self.max_epochs:
            
            count_e = 0
            for x_i, d_i in zip(xs, ds):
                ws = [] 
                
                # weighted sum of features
                z = self._weighted_sum(xs=x_i, ws=WS[-1])
                sums.append(z)
                
                # activation  function
                if z<self.threshold:
                    y=0
                else:
                    y=1
                
                # actualization of the weights
                err  = d_i - y
                
                for k in range(self.cols):
                    ws.append(WS[-1][k] + self.alpha*x_i[k]*err)
                    
                WS.append(ws)  
                
                # if the weights are not upgrated in a full epoch then 
                # the algorithm stops
                count_e += abs(err)
                
            errs.append(count_e)
            epochs+=1
            if count_e==0: break
            
        self.epochs = epochs
        self.ws     = WS[-1]
        if count_e==0:
            self.succes="yes"
        else:
            self.succes="no"
        
        if self.epochs==self.max_epochs and self.succes=="no":
            import warnings
            warnings.warn(f"Max number of epochs {self.max_epochs} reached without convergency, err={count_e}")
            
        return {"full_weights": WS, "final_weights":WS[-1], "errors":errs, "weighted_sums":sums}
        
    
    #====================================================================#
    #                       proof of the perceptron                      #
    #====================================================================#
    def proof_perceptron(self):
    
        title = ""
        for i in range(len(self.xs[0])):
            title += f"x_{i+1}\t"

        title += "d\ty"
        print(title)

        for row, d in zip(self.xs, self.ds):
            z = self._weighted_sum(row, self.ws)

            #Appliying stepwise activation function
            if z<self.threshold: y = 0 
            else:y = 1

            line = ""
            for i in range(len(self.ws)):
                line+=f" {row[i]}\t"
            line+=f"{d}\t{y}"

            print(line)
            
    #====================================================================#
    #                              predict                               #
    #====================================================================#
    def predict(self, xs):
        pred = []
        for x_i in xs:
            z = self._weighted_sum(xs=x_i, ws=self.ws)
            if z<self.threshold: pred.append(0)
            else:pred.append(1)
        
        return pred
        
 
    #====================================================================#
    #                          to string method                          #
    #====================================================================#
    def __str__(self):
        return f"\nalpha (learning ratio): {self.alpha}"+\
               f"\nthreshold: {self.threshold}"+\
               f"\nmax_epochs: {self.max_epochs}"+\
               f"\nepochs: {self.epochs}"+\
               f"\nrandom state: {self.random_state}"+\
               f"\nsucces: {self.succes}"
```

```python
xs_and  = (1,1), (1,0), (0,1), (0,0)
ds_and  = 1,0,0,0
ws_and0 = "auto"

p = Perceptron()
print(p)
```

```python
sol = p.perceptron(xs=xs_and, ds=ds_and, ws=ws_and0)
```

### Diferenciado entre clases: frontera de decisión

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
# generando las clases
xs_blobs, ys_blobs = make_blobs(n_samples=1000,
                                n_features=2,
                                centers=((8,-2), (6,6)),
                                shuffle=True, random_state=2)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.scatter(x=xs_blobs[:,0], y=xs_blobs[:,1],
           c=ys_blobs,
           s=100, alpha=0.5)
ax.grid()
plt.show()
```

```python
p_blobs   = Perceptron(max_epochs=100, random_state=5)
sol_blobs = p_blobs.perceptron(xs=xs_blobs,
                               ds=ys_blobs,
                               ws="random")

print(p_blobs)
```

```python
#========================================================#
# frontera de decisión
xs = np.linspace(min(xs_blobs[:,0]),
                 max(xs_blobs[:,0]),
                 100)

ys = (0.5 - xs*sol_blobs["final_weights"][0])/sol_blobs["final_weights"][1]

fig, ax = plt.subplots(figsize=(8,8))
#ax.plot(xs, ys, c="k", lw=2)

#========================================================#
# clases
scatter = ax.scatter(x=xs_blobs[:,0], y=xs_blobs[:,1],
                     edgecolors="black",
                     c=ys_blobs,   s=100, alpha=0.5)

#========================================================#
# plot regiones
xx, yy = np.meshgrid(np.linspace(xs_blobs[:,0].min()-2, xs_blobs[:,0].max()+2, 100),
                     np.linspace(xs_blobs[:,1].min()-2, xs_blobs[:,1].max()+2, 100)
                    )

zz = np.array(p_blobs.predict(np.c_[xx.flatten(), yy.flatten()])).reshape(xx.shape)

# regiones
CS = ax.contourf(xx, yy, zz,
                 alpha=0.25)

# frontera
CS = ax.contour(xx, yy, zz,
                    levels=[0.5],
                    linestyles=["-"],
                    colors=["k"], linewidths=2 
                    )

ax.clabel(CS,
          fontsize=20,
          manual=[(10,4)]) # Labels para las superficies de nivel.

#========================================================#
# custum
fontl = {'family' : 'DejaVu Sans',
            'weight' : 'bold',
            'size'   : 12}

font = {'family' : 'DejaVu Sans',
            'weight' : 'bold',
            'size'   : 18}

ax.legend(*scatter.legend_elements(num=np.ndarray.tolist(np.unique(ys_blobs))),
           ncol=1,
           loc="best",
           frameon=False, prop=fontl)

ax.set_xlim(min(xs), max(xs))
ax.set_title(f"Perceptron\n$w_0={sol_blobs['final_weights'][0]:.4f}$\n$w_1={sol_blobs['final_weights'][1]:.4f}$\n",
             fontdict=font)
ax.set_xlabel("feature 1", **font)
ax.set_ylabel("feature 2", **font)
plt.grid()
plt.show()
```

## Usando <code>perceptron</code> de Sklearn.

La librería Sklearn nos permite implementar la clase <code>perceptron</code> del módulo <code>lienar_model</code> que es lo mismo que la implementación de <code>SGDClassifier</code> utilizando como hiperparámetros <code>loss="perceptron"</code> y <code>learning_rate="constant"</code>. 

```python
from sklearn.linear_model import Perceptron
from sklearn import metrics
```

```python
psk_blobs = Perceptron(alpha=0.001, random_state=5)
psk_blobs.fit(X=xs_blobs, y=ys_blobs)
```

```python
psk_blobs.coef_
```

```python
psk_blobs.intercept_
```

```python
#========================================================#
# frontera de decisión
xs = np.linspace(min(xs_blobs[:,0]),
                 max(xs_blobs[:,0]),
                 100)

ys = (0.5 - xs*psk_blobs.coef_[0][0])/psk_blobs.coef_[0][1]

fig, ax = plt.subplots(figsize=(8,8))
#ax.plot(xs, ys, c="k", lw=2)

#========================================================#
# clases
scatter = ax.scatter(x=xs_blobs[:,0], y=xs_blobs[:,1],
                     edgecolors="black",
                     c=ys_blobs,   s=100, alpha=0.5)

#========================================================#
# plot regiones
xx, yy = np.meshgrid(np.linspace(xs_blobs[:,0].min()-2, xs_blobs[:,0].max()+2, 100),
                     np.linspace(xs_blobs[:,1].min()-2, xs_blobs[:,1].max()+2, 100)
                    )

zz = np.array(psk_blobs.predict(np.c_[xx.flatten(), yy.flatten()])).reshape(xx.shape)

# regiones
CS = ax.contourf(xx, yy, zz,
                 alpha=0.25)

# frontera
CS = ax.contour(xx, yy, zz,
                    levels=[0.5],
                    linestyles=["-"],
                    colors=["k"], linewidths=2 
                    )

ax.clabel(CS,
          fontsize=20,
          manual=[(10,4)]) # Labels para las superficies de nivel.

#========================================================#
# custum
fontl = {'family' : 'DejaVu Sans',
            'weight' : 'bold',
            'size'   : 12}

font = {'family' : 'DejaVu Sans',
            'weight' : 'bold',
            'size'   : 18}

ax.legend(*scatter.legend_elements(num=np.ndarray.tolist(np.unique(ys_blobs))),
           ncol=1,
           loc="best",
           frameon=False, prop=fontl)

ax.set_xlim(min(xs), max(xs))
ax.set_title(f"Perceptron\n$w_0={psk_blobs.coef_[0][0]:.4f}$\n$w_1={psk_blobs.coef_[0][1]:.4f}$\n",
             fontdict=font)
ax.set_xlabel("feature 1", **font)
ax.set_ylabel("feature 2", **font)
plt.grid()
plt.show()
```
